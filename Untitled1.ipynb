<h3> b)Write a Program to implement Static and Instance methods,<br> Abstract Classes and Interfaces. </h3> 

<h4> Static and Instance method </h4>

```python
class captain:
  def __init__(self,name,winyear):
    self.name = name
    self.winyear = winyear


  #Instance method
  def displayData(self):
    print("Name of the Captain:",self.name)
    print("winning year of the captain:",self.winyear)

  #Static method
  def captainsData(captain):
    print()
    print("-"*30)
    for cap in captain:
      cap.displayData()
      print("-"*30)

s_1 = captain("Kapil Dev",1983)
s_2 = captain("Ms Dhoni",2011)
s_3 = captain("Sourav Ganguly",2004)

s_1.displayData()
s_2.displayData()
s_3.displayData()

captain.captainsData([s_1,s_2,s_3])
```

<h4>Abstract Class</h4>

```python
from abc import ABC, abstractmethod

class DL(ABC):
  @abstractmethod
  def desc(self):
    pass

class ANN(DL):
  def desc(self):
    print("ANN is a technique which use for classification and regression")

class CNN(DL):
  def desc(self):
    print("CNN is a technique which is used for Image data")

class RNN(DL):
  def desc(self):
    print("RNN is a technique which is mainly used in text data")
d1 = ANN()
d1.desc()

d2 = CNN()
d2.desc()

d3 = RNN()
d3.desc()

  
```

a) Write a program to implement Inheritance and Polymorphism
</h3>

```python
#Parent class
class captain:
  def __init__(self,name,year):
    self.name = name
    self.year = year

  def captaindata(self):
    print()
    print("-"*30)
    print("Name of the captain:",self.name)
    print("winning year of the captain:",self.year)

#child class
class previousworldcup(captain):
  def __init__(self,name,year,prevcup):
    captain.__init__(self,name,year)
    self.prevcup = prevcup

#child class
class currcup(captain):
  def __init__(self,name,year,currcup):
    captain.__init__(self,name,year)
    self.currcup = currcup

stud1 = previousworldcup("Kapil Dev",1983,"ICC CRICKET WORLD CUP 1983")
stud1.captaindata()
print(stud1.prevcup)

stud2 = currcup("Ms Dhoni",2011,"ICC CRICKET WORLD CUP 2011")
stud2.captaindata()
print(stud2.currcup)
```

<h3>
c) Program to implement perceptron
</h3>

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt


class Perceptron():
    def __init__(self,lr=0.01,niter=20):
        self.lr = lr
        self.niter = niter 
        
    def fit(self,X,y):
        self.w_ = np.zeros(1+X.shape[1])
        self.errors_ = []
        h = 0
        for _ in range(self.niter):
            errors = 0
            for xi, target in zip(X,y):
                update = self.lr*(target-self.predict(xi))
                self.w_[1:]+=update*xi
                self.w_[0]+=update
                errors+=int(update != 0.0)
                h+=1
            self.errors_.append(errors)
        print("value of h: ",h)
        return self
    
    def net_input(self,X):
        return np.dot(X,self.w_[1:])+self.w_[0]
    
    def predict(self,X):
        return np.where(self.net_input(X)>=0.0,1,-1)
```

```python
df = pd.read_csv("/iris.csv")
y = df.iloc[0:100,4].values
y = np.where(y == 'Iris-setosa',-1,1)
X = df.iloc[0:100,[0,2]].values
```

```python
plt.scatter(X[:50,0],X[:50,1],color='black',marker='+',label='setosa')
plt.scatter(X[50:100,0],X[50:100,1],color='blue',marker='o',label='versicolor')
plt.xlabel('petal length')
plt.ylabel('sepal length')
plt.legend(loc='upper left')
plt.show()
```

```python
per = Perceptron(lr=0.01, niter=20)
per.fit(X,y)
```

```python
plt.plot(range(1, len(per.errors_)+1), per.errors_,marker='o',color='red')
plt.xlabel('Epochs')
plt.ylabel("Number of misclassifications")
```

a) Generate a decision tree. Find the Depth of decision trees and observe the<br> results, then
propose some changes in DecisionTreeClassifier function to limit.
</h3>

```python
from sklearn import tree 
X = [[0, 0, 0], [5, 5, 5], [3, 3, 3]] 
Y = [0,0,1] 
clf = tree.DecisionTreeClassifier() 
clf = clf.fit(X, Y) 
print(clf.tree_.max_depth)
```

```python
class Node:
    def __init__(self, data):
        self.data = data 
        self.left = None 
        self.right = None 
    
def maxDepth(node): 
    if node is None: 
        return 0; 
    else: 
        lDepth = maxDepth(node.left) 
        rDepth = maxDepth(node.right) 
        if (lDepth > rDepth): 
            return lDepth+1 
        else: 
            return rDepth+1 
            
root = Node(30) 
root.left = Node(31) 
root.right = Node(38) 
root.left.left = Node(25) 
root.left.right = Node(43) 
root.left.left.left = Node(11) 
print("Height of tree is %d" % (maxDepth(root)))
```
